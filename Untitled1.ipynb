```python
numeric_columns = churned_df.select_dtypes(include=[float, int])

# Create a PCA instance with the same number of components as columns in 'numeric_columns'
pca = PCA(n_components=len(numeric_columns.columns))

# Fit the PCA model to your data
pca.fit(numeric_columns)

# Normalize the numeric columns using the mean and standard deviation calculated during PCA
df_normalized = (numeric_columns - pca.mean_) / np.sqrt(pca.explained_variance_)

# Calculate the covariance matrix
c_matrix = np.dot(df_normalized.T, df_normalized)

#  eigenvalues
eigenvalues = [np.dot(eigenvector.T, np.dot(c_matrix, eigenvector)) for eigenvector in pca.components_]
```

```python
plt.plot(eigenvalues)
plt.xlabel('number of components')
plt.ylabel('eigenvalue')
plt.show() 
```

```python
loadings = pd.DataFrame(pca.components_.T[:, :3],  # Only select the first 3 components
                       columns=['PC1','PC2','PC3'],
                       index=numeric_columns.columns)
print(loadings)
```
